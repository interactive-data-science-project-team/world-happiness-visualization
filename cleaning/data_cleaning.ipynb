```python
cleaned_data = data.groupby('country', as_index=False).fillna(method='ffill')
cleaned_data = cleaned_data.groupby('country', as_index=False).fillna(method='bfill')
cleaned_data
```

```python
# not sufficient data before 2007
cleaned_data = cleaned_data[cleaned_data.year >= 2007]
```

```python
cleaned_data = cleaned_data.drop('GINI index (World Bank estimate)', axis=1)
```

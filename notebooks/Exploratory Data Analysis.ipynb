```python
happiest2017 = df.loc[df['year'] == 2017].sort_values('Happiness Score', ascending=False)[0:5]['country'].values
mask = df.country.map(lambda c: c in happiest2017)
```

```python
plt.figure(figsize=(12, 6))

sns.boxplot(x='year', y='Happiness Score',data=df)
plt.ylabel('Happiness Score')

plt.show()
```

```python
plt.figure(figsize=(12, 24))

for i, y in enumerate(['Happiness Score', 'Log GDP per capita', 'Healthy life expectancy at birth', 'Democratic Quality']):
    plt.subplot(4, 1, i + 1)
    sns.boxplot(x='year', y=y, data=df)
    # plt.ylabel('Log GDP per capita')

plt.show()
```

```python
plt.figure(figsize=(18, 6))
sns.set()
sns.set_style('white')
sns.violinplot(x='year', y='Healthy life expectancy at birth',data=df)
plt.ylabel('Healthy life expectancy at birth')

plt.show()
```

```python
plt.figure(figsize=(12, 6))


happiest2017 = df.loc[df['year'] == 2017].sort_values('Happiness Score', ascending=False)[0:10]['country'].values
mask = df.country.map(lambda c: c in happiest2017) & (df['year'] == 2017)

data = df[mask].sort_values('Happiness Score', ascending=False)
data.country = data.country.map(lambda c: c[0:10])

sns.barplot(x='country', y='Happiness Score',data=data)
plt.ylabel('2017 Happiness Score')
plt.ylim((0, 8))

plt.show()
```

```python
# unhappiest countries

plt.figure(figsize=(12, 6))

happiest2017 = df.loc[df['year'] == 2017].sort_values('Happiness Score', ascending=True)[0:10]['country'].values
mask = df.country.map(lambda c: c in happiest2017) & (df['year'] == 2017)

data = df[mask].sort_values('Happiness Score', ascending=True)
data['country'] = data.country.map(lambda c: c[0:10])
sns.barplot(x='country', y='Happiness Score', data=data)
plt.ylabel('2017 Happiness Score')
plt.ylim((0, 8))

plt.show()
```
